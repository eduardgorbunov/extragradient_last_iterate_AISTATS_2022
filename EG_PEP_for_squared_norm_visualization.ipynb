# EG: analysis of $\|F(x^N)\|^2$

## $L = 1, \mu = 0$

```python
L = 1.0
gamma1 = 1.0/(2*L)
gamma2 = 1.0/(2*L)
N_max = 50

worst_case_sq_norm = np.array([])
guessed_upper_bound = np.array([])

gamma1_str = '%.6f' % gamma1
gamma2_str = '%.6f' % gamma2

for N in range(N_max):
    res = scipy.io.loadmat('dump/EG_norm_L_'+str(int(L))+'_N_'+str(N+1)+'__'+gamma1_str+'_'+gamma2_str+'.mat')
    worst_case_sq_norm = np.append(worst_case_sq_norm, res['res_norm'][0][0])
    guessed_upper_bound = np.append(guessed_upper_bound, 16.0*L*res['res_init_dist'][0][0]/(3*(N+1)))
    
fig = plt.figure(figsize=(12,8))
plt.title(r"Worst-case $||F(x^k)||^2$ for EG, $L = 1$, $\gamma_1 = \gamma_2 = \gamma = \frac{1}{2L}$", fontsize=28)

plt.semilogy(np.array(range(N_max))+1, worst_case_sq_norm, linewidth=2,
             markersize = 12, marker = "o",
             markevery=range(0, 50, 10), color="forestgreen",
             label='PEP answer')
plt.semilogy(np.array(range(N_max))+1, guessed_upper_bound, linewidth=2, 
             markersize = 12, marker = "d",
             markevery=range(0, 50, 10), color="crimson",
             label=r'$\frac{16L^2||x^0 - x^*||^2}{3k}$')

# plt.plot(np.array(range(N_max))+1, worst_case_sq_norm, linewidth=2, label='PEP answer')
# plt.plot(np.array(range(N_max))+1, guessed_upper_bound, linewidth=2, label=r'$\frac{L\|x^0 - x^*\|^2}{\gamma k}$')

plt.legend(loc='upper right', fontsize=25)
plt.xlabel(r"$k$", fontsize=30)
plt.ylabel(r"$||F(x^k)||^2$", fontsize=30)

plt.xticks(fontsize=25)
_ = plt.yticks(fontsize=25)

plt.grid()

ax = fig.gca()
ax.xaxis.offsetText.set_fontsize(28)
ax.yaxis.offsetText.set_fontsize(28)

# ax.set_xscale('log')

plt.savefig("plot/EG_norm_L_"+str(L)+'_N_'+str(N_max)+".pdf", bbox_inches='tight')
```
